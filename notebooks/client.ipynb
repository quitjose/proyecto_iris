```python
import requests
import json
import random

def enviar_solicitudes(modelos, repeticiones=2):
    for _ in range(repeticiones): 

        datos_aleatorios = {
            "PetalLengthCm": random.uniform(0.1, 10.0),
            "PetalWidthCm": random.uniform(0.1, 5.0)
        }


        for modelo in modelos:
            print(f"Enviando solicitudes para el modelo {modelo}...")
            print("Datos enviados:")
            print(f"petal_length: {datos_aleatorios['PetalLengthCm']}")
            print(f"petal_width: {datos_aleatorios['PetalWidthCm']}")

            # Utiliza los mismos datos aleatorios para cada modelo
            parametros = {"PetalLengthCm": datos_aleatorios["PetalLengthCm"],
                          "PetalWidthCm": datos_aleatorios["PetalWidthCm"],
                          "Modelo": modelo}

            datos_json = json.dumps(parametros)

            respuesta = requests.post(url, data=datos_json, headers=cabeceras)

            print(respuesta.text)

# Lista de modelos
modelos = ["regresion_model", "svm_model", "tree_model", "knn_model"]

url = "http://127.0.0.1:8000/predict"
cabeceras = {"Content-Type": "application/json"}


enviar_solicitudes(modelos, repeticiones=2)
```
